```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../dataset/reg_sample.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
# visualisasi data kolom x dan y
plt.scatter(df['x'], df['y'])
plt.show()
```

```python
# pisahkan dataframe ke variabel x dan y
x = df['x'].values.reshape(-1, 1)
y = df['y'].values.reshape(-1, 1)
```

```python
# print nilai x
x
```

```python
# print nilai y
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# training data ke model regresi linear dengan metode OLS/bentuk model regresi
reg = LinearRegression().fit(x, y)
```

```python
# visualisasi model regresi
x1 = np.array([x.min(), x.max()]).reshape(-1, 1)

plt.plot(x1.ravel(), reg.predict(x1).ravel(), color='red', linestyle='--')
plt.scatter(df['x'], df['y'])
plt.show()
```

```python
# koef regresi (b)
reg.coef_
```

```python
# konstanta regresi (a)
reg.intercept_
```

```python
# R-squared
reg.score(x, y)
```

```python
y_pred = reg.predict(x) 
y_pred
```

```python
from sklearn.metrics import mean_squared_error as mse
```

```python
# MSE
mse(y, y_pred)
```

```python
# RMSE
np.sqrt(mse(y, y_pred))
```
